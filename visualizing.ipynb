```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
outputs = {0.001: 0.9480, 0.01: 0.9481, 0.1: 0.9454, 1.0: 0.9024, 10.0: 0.5810}

colors = sns.color_palette('colorblind')

plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(outputs)), list(outputs.values()), color=colors)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.4f}', ha='center', va='bottom')

plt.xticks(range(len(outputs)), list(outputs.keys()))
plt.xlabel('Alpha Value')
plt.ylabel('Accuracy')
plt.title('Development Set Accuracy vs. Alpha Value')
plt.show()
```
